# 2.Crible d’ÉRATOSTHÈNE:  
RAPPELS : Un nombre premier est un nombre entier positif (supérieur à 1) qui se trouve être divisible uniquement par lui-même (et par 1). Par exemple : 2, 3, 5, 7, 11, 13 ... Donc 4 n'est pas un nombre premier car il est divisible par 2 (2x2=4)

Pour vérifier si un nombre N est un nombre premier, on pourrait essayer de diviser celui-ci par tous les nombres entiers <N (et si on ne trouve pas de diviseurs, on dit qu'il est premier), mais cela prendrait trop de temps pour trouver tous les nombres premiers inférieurs à 1000 ou à 10000.

ÉRATOSTHÈNE a établi une méthode beaucoup plus efficace et plus rapide pour trouver tous les nombres premiers inférieurs à une borne donnée :

On écrit tous les nombres jusqu’à MAX (dans une liste par exemple : [2, 3, 4, ... ,MAX]).

On balaye ensuite la liste de gauche à droite.

Si un nombre n’est pas « barré », c’est un nombre premier. Dans ce cas, on « barre » tous ses multiples jusqu’à MAX.

Puis on teste le suivant, etc. jusqu’à MAX.

Les nombres restants sont premiers ! Il ne reste plus qu’à les afficher


1. Crée une liste contenant les valeurs de 2 jusqu’à MAX inclus.

```python
#Ajout d'un input pour le max d'une liste
m=int(input('donner max:'))

#creation d'une liste
# pour x dans l'intervalle 2 a max 
for x in range (2,m):
    # si x est divisible par 2 a pour reste 0
    if x%2==0:
        # Afficher x
        print(x,end=' ')
    # Incrementation de x +1
    x+=1
```

2. Écrire ensuite une fonction qui réalise le  traitement de la  liste  conformément à la  méthode du crible
d’ÉRATOSTHÈNE. On représentera un nombre barré par le fait de remplacer sa valeur par zéro dans la liste.

```python
#Ajout d'un input pour le max d'une liste
m=int(input('donner max:'))

# m=20
#Ajout de 1 pour pouvoir avoir la valeur max de max
m+=1
#intialisation de la variable j au format liste
j=[]

# creation de la liste
# pour x dans l'intervalle 2 a max 
for x in range (2,m):
    # Ajout de x a j
    j.append(x)

# debugage print(j)

#Methode Eratosthene
# pour x dans l'intervalle 0 a la taille de la liste : 
for x in range (0,len(j)):
    # si la valeur de x dans j est egales a 0 ou 1 
    if j[x]==0 or j[x]==1:
        # Continuer (continuer la boucle)
        continue
    
    # Pour y dans l'intervalle x+1 , taille de la liste :
    for y in range (x+1,len(j)):
        # Si la valeur de y dans j quand divise par la valeur de x dans j a pour reste 0  
        if j[y]%j[x]==0:
            # Assigner a cette position 0 
            j[y]=0
        
# affichage de la liste j avec les modification
print(j)
```

3. Écrire une fonction qui permette d’afficher tous les éléments non nuls d’une liste (ce qui reviendra à afficher
tous les nombres premiers inférieurs à MAX d’une liste traitée).


```python
# Nouvelle liste nommer nl
nl = []
# pour x dans j:
for x in j:
    # si x est different de 0
    if x != 0:
        # Ajouter x a nl
        nl.append(x)
# filtre au niveau nl pour les valeurs vides
nl=list(filter(None,nl))
# Affichage de nl
print(nl)
```
