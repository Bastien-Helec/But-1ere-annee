````python
import dns.resolver

class Resolveur: 
    """Class custom resolver ns : IP du dns qui fera la résolution"""
    def __init__(self,ns : str = '1.1.1.1') -> str:
        self.ns = ns

    def resoudFQDN(self, mon_fqdn, servertype):
        """Prend un fqdn et retourne une instance de IPAddress(NetAddr)"""
        Fqdn= dns.resolver.resolve(mon_fqdn, servertype)
        for data in Fqdn:
            # print(f"host {data}")
            return data
        


    def resoudInverse(self, mon_ipaddr)-> str:
        """Prend une instance de IPAddress(NetAddr) et retourne un FQDN"""
        ipaddr= dns.resolver.resolve_address(mon_ipaddr)
        for data in ipaddr:
            # print(f"host {data}")
            return data


mon_resolveur = Resolveur().resoudFQDN('iutbeziers.fr', 'NS')
mon_resolveur_inverse= Resolveur().resoudInverse('194.199.227.80')

print(mon_resolveur, mon_resolveur_inverse)
````

````python
import nmap
import netaddr
import logging


targets:tuple[str,...] = ('172.24.102.237','10.255.255.200')
dports:tuple[int,...] = (21,80,443,161,3306,4443,8080)
logging.getLogger("nmap").setLevel(logging.ERROR)

# create a port scanner object
scanner = nmap.PortScanner()

# scan the target ip and ports
for ip in targets:
    print("scan de l'IP: {}\n".format(ip))
    for port in dports:
        print(f"scan de du port: {port}\n")
        try:
            scanner.scan(ip, str(port))
            state=scanner[ip]['tcp'][port]['state']
            print("Port {}: {}".format(port, state))
            print(50*"=")
        except KeyError as e:            
            print("machine inaccessible")
            break
        
````
