8)- Ecrire un programme qui permet de récupérer l’occupation du parking « FR_MTP_GARE » toutes
les 10 secondes pendant 5 minutes et qui sauvegarde ces données dans un fichier

```python

import requests
from lxml import etree
import time
from time import sleep
def times():
    response=requests.get("https://data.montpellier3m.fr/sites/default/files/ressources/FR_MTP_GARE.xml")
    print(response.text)

    f1=open("FR_MTP_GARE.txt","a", encoding='utf8')

    f1.write(response.text)
    f1.close()

    tree = etree.parse("FR_MTP_GARE.txt")

    f6=open('trie3.txt','a',encoding='utf8')

    for user in tree.xpath("Total"):
        print('Nombre total de places :',user.text)
        f6.write(user.text+'\n')
    for user in tree.xpath("Free"):
        print('Nombre de places libres :',user.text)
        f6.write(user.text+'\n')
    temps=int(time.time())
    print(temps)
    f6.write('temps :'+str(temps)+'\n')
    f6.close()

#Toutes les 10 secondes on récupère les données pendant 5 minutes avec sleep

for i in range(30):
    times()
    sleep(10)  


```
